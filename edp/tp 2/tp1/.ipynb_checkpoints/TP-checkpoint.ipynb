## 1 Construction et visualisation du maillage et de la condition initiale

```python
L = 1
m = 50
x0 = 0.5
delta_x = L/(m-1)
```

```python
X = np.linspace(start=0, stop=L, num=m)
X2 = np.array([i*(L/(m-1)) for i in range(m)])
```

```python
def CI1(X, beta=2):
    u_zero = np.sin((2*beta*X*np.pi)/L)
    return u_zero
```

```python
u_zero = CI1(X, beta=2)
plt.plot(u_zero)
plt.show()
```

```python
def CI2(X, x0=0.5):
    u_zero = np.array([(0 if xi<x0 else 1) for xi in X])
    return u_zero
```

```python
u_zero = CI2(X, x0=0.5)
plt.plot(u_zero)
plt.show()
```

## 2 Etude du comportement de schémas différences finies

### 2.1 Paramètres physiques et numériques

### 2.2 Méthodologie

```python
def get_u_i(u, i):
    I = u.shape[0]
    if 0 <= i < I:
        return u[i]
    else:
        return get_u_i(u, i%I)
```

```python
u_zero
```

```python
np.roll(u_zero, shift=+1)
```

```python
def FOU_EE(ui, C=0.5):
    u = ui - C*(ui - np.roll(ui, shift=+1))
    return u
```

```python
def FOF_EE(ui, C=0.5):
    u = ui - C*(np.roll(ui, shift=-1) - ui)
    return u
```

```python
def SOC_EE(ui, C=0.5):
    u = ui - C*(np.roll(ui, shift=-1) - np.roll(ui, shift=+1))
    return u
```

```python
def Lax_Wendroff(ui, C=0.5):
    u = ui - 0.5*C*(np.roll(ui, shift=-1) - np.roll(ui, shift=+1)) + \
        0.5*C*(np.roll(ui, shift=-1) - 2*ui + np.roll(ui, shift=+1))
    return u
```

### DEMO

```python
def get_X(L=1, m=50):
    L = 1
    m = 50
    X = np.linspace(start=0, stop=L, num=m)
    return X
```

```python
def plot_sol(schema=Lax_Wendroff, fonction_name='CI1', L=1, m=50, C=1.0, a=1.0, beta=1, x0=1.0, P=1):
    X = get_X(L, m)
    if fonction_name == "CI1":
        fonction = CI1
        u0 = fonction(X, beta)
    else:
        fonction = CI2
        u0 = fonction(X, x0)
    delta_x = L/(m-1)
    delta_t = C*delta_x/a
    tMax = int(1/delta_t)
    tMax = int(tMax/P)

    # t = 0
    ut = u0
    plt.plot(X, ut)

    # t inter
    for t in range(tMax):
        ut = schema(ut, C)
        plt.plot(X, ut, c=(t/tMax, 0., 0.), alpha=0.1)

    # t = tend
    ut = schema(ut, C)
    plt.plot(X, ut)

    plt.title(f"{fonction.__name__}, {schema.__name__}, beta={beta}, L={L}, m={m}, C={C}, a={a}, Tmax{P}")
    plt.savefig(
        f"{fonction.__name__}__{schema.__name__}__beta={beta}__L={L}__m={m}__C={C}__a={a}__Tmax{P}.png")
    plt.show()
```

```python
for beta in [2, 4, 8, 16]:
    plot_sol(schema=FOU_EE, fonction_name='CI1', L=1,
             m=50, C=1.0, a=1.0, beta=beta, x0=1.0, P=1)

for C in [0.1, 0.2, 0.4, 0.8, 1, 1.5]:
    plot_sol(schema=FOU_EE, fonction_name='CI1', L=1,
             m=50, C=C, a=1.0, beta=1.0, x0=1.0, P=1)

for P in [8, 4, 2, 1]:
    plot_sol(schema=FOU_EE, fonction_name='CI2', L=1,
             m=50, C=1.0, a=1.0, beta=1.0, x0=1.0, P=P)

for C in [0.1, 0.2, 0.4, 0.8, 1, 1.5]:
    plot_sol(schema=FOU_EE, fonction_name='CI2', L=1,
             m=50, C=C, a=1.0, beta=1.0, x0=1.0, P=1)

for m in [4, 10, 20, 50, 100, 200]:
    plot_sol(schema=FOU_EE, fonction_name='CI2', L=1,
             m=m, C=0.5, a=1.0, beta=1.0, x0=1.0, P=1)
```
