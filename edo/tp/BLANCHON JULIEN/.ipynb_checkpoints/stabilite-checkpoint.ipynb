# TP Stabilité

```julia
using Plots
```

```julia
A1 = [-1 -4; 1 -1]
A2 = [1 -4; 1 -1]
A3 = [1 1; 4 -2]

tf1 = 2*pi/√3
tf2 = 1
```

```julia
function trajectoire(A, t0, tf, nX, nT, r0, x0, y0)
    θ1 = LinRange(0, 2*π, nX);
    X1 = x0 .+ r0*cos.(θ1);
    Y1 = y0 .+ r0*sin.(θ1);
    U = [X1 Y1]
    
    T = LinRange(t0, tf, nT);
    
    plt = plot()
    for ix in 1:nX
        Xi = U[ix, :] #(2,)
        XiT = zeros(nT, 2) #(nT, 2)
        for it in 1:nT
            XiT[it, :] = exp(T[it]*A)*Xi
        end
        plt = plot!(XiT[:, 1], XiT[:, 2], legend=false, color=:blue)
    end
end
```

```julia
function flot(A, t0, tf, nX, nT, r0, x0, y0, plt)
    θ1 = LinRange(0, 2*π, nX);
    X1 = x0 .+ r0*cos.(θ1);
    Y1 = y0 .+ r0*sin.(θ1);
    U = [X1 Y1]
    
    T = LinRange(t0, tf, nT);
    
    for it in 1:nT
        Xt = zeros(nX, 2)
        for ix in 1:nX
            Xt[ix, :] = exp((T[it]-t0)*A)*U[ix, :]
        end
        plt = plot!(Xt[:, 1], Xt[:, 2], legend=false, color=:green, fillrange = 0, fillalpha = 0.5, fillcolor = :green)
    end
end
```

```julia
plt = plot()
trajectoire(A1, 0, tf1, 10, 100, 0.3, 0.8, 0.8, plt)
flot(A1, 0, tf1, 30, 9, 0.3, 0.8, 0.8, plt)
display(plt)
```

```julia
plt = plot()
trajectoire(A2, 0, tf1, 10, 100, 0.3, 0.8, 0.8, plt)
flot(A2, 0, tf1, 30, 9, 0.3, 0.8, 0.8, plt)
display(plt)
```

```julia
plt = plot()
trajectoire(A3, 0, tf2, 10, 100, 0.1, 0.15, -0.6, plt)
flot(A3, 0, tf2, 30, 4, 0.1, 0.15, -0.6, plt)
display(plt)
```
