## Introduction

Il est rappelé que les programmes doivent respecter les **bonnes pratiques de la programmation**. En particulier on vérifiera que les interfaces soient bien définies (paramètres en entrée, en sortie avec leurs types, les dimensions, ...). Dans le cas contraire on mettra des **points négatifs** pour un maximum de 4 points.

On rappelle qu'un schéma de Runge Kutta à $s$ étage est défini par
\begin{equation}\label{eq:irk}
(S)\left\{\begin{array}{l}
k_i=f(t_0+c_ih,x_0+h\sum_{j=1}^{s}a_{ij}k_j)\quad \textrm{pour}\;\;i=1,\ldots,s\\
x_1=x_0+h\sum_{i=1}^{s}b_ik_i
\end{array}\right.
\end{equation}
où les coefficients $c_i,a_{ij}$ et $b_i$ sont des constantes qui définissent précisément le schéma. On supposera toujours dans la suite que  $c_i=\sum_{j=1}^{s}a_{ij}$ pour $i=1,\ldots,s$. 

On représente en pratique ce schéma par le tableau de Butcher
$$
\begin{array}{c|ccc}
c_1    & a_{11} & \ldots & a_{1s}\\
\vdots & \vdots &        & \vdots\\ 
c_s    & a_{s1} & \ldots & a_{ss}   \\ \hline
       & b_1    & \ldots & b_s\\
\end{array}$$

Si la matrice des coefficient $A$ est triangulaire inférieure stricte alors le schéma est explicite, sinon il est implicite.

On donne ci après les schémas les plus courant

$$\begin{array}[t]{ccc}
\begin{array}{c}
\\ \\
\begin{array}{c|c}
0 &   \\\hline
 & 1
\end{array}
\end{array}
& \begin{array}{c}
\\
\begin{array}{c|cc}
0 &  &\\
1/2 & 1/2 &\\ \hline
 & 0 & 1
\end{array}
\end{array}
& \begin{array}{c|ccc}
0 &  &  & \\
1/3 & 1/3 & & \\
2/3 & 0 & 2/3 & \\ \hline
 & 1/4 & 0 & 3/4
\end{array}\\
\textrm{Euler} & \textrm{Runge} & \textrm{Heun}
\end{array}$$
$$\begin{array}[t]{cc}
\begin{array}{c|cccc}
0 &  &  & &\\
1/2 & 1/2 & & &\\
1/2 & 0 & 1/2 & &\\ 
1 &   0 & 0   & 1 &\\ \hline
 & 1/6 & 2/6 & 2/6 & 1/6
\end{array}
& \begin{array}{c|cc}
1/2-\sqrt{3}/6 & 1/4 & 1/4-\sqrt{3}/6\\
1/2+\sqrt{3}/6 & 1/4+\sqrt{3}/6 & 1/4\\ \hline
 & 1/2 & 1/2
\end{array}\\
\textrm{La méthode Rk4} & \textrm{Gauss d'ordre 4}
\end{array}$$

## Les méthodes de Runge Kutta explicites

```julia
function euler(fun, tspan::Tuple{<:Real,<:Real}, x0::Vector{<:Real}, p::Vector{<:Real},
        N::Integer)::Tuple{Vector{Float64},Array{Float64,2}}
# Numerical integration via Euler scheme
# Input
# -----
# fun : second member of the ode
#       xpoint = fun(x,p,t)
#       xpoint : value of fun
#               array(n)
#       x      : state
#                array(n)
#       p      : vector of parameter
#       t      : time
#                real
# tspan : (t0, tf)
# x0 : initial condition
#       array(n)
# p  : vector of parameters
# N  : number of step
#      integer
# Output
# ------
# T   : vector of time
#       array(N+1)
# X   : X[i,:] is the approximation of x(t_i)
#       array(N+1,n)
# ----------------------------------------------
    n  = length(x0)             # dimension n
    T  = zeros(N+1)
    X  = zeros(N+1,n)
    h = ((tspan[2]-tspan[1])/N) # step
    
    #Filling X and T
    T[1] = tspan[1]
    X[1,:] = x0
    for step in 2:N+1
        T[step] = T[step-1] + h
        X[step,:] = X[step-1,:] + h * fun(X[step-1,:], p, T[step-1])
    end
    
    return T,X
end
```

```julia
function runge(fun, tspan::Tuple{<:Real,<:Real}, x0::Vector{<:Real}, p::Vector{<:Real}, N::Integer)
# Numerical integration via the runge scheme
# the same as Euler
# ----------------------------------------------

    n  = length(x0)             # dimension n
    T  = zeros(N+1)
    X  = zeros(N+1,n)
    h = ((tspan[2]-tspan[1])/N) # step
    
    #Filling X and T
    T[1] = tspan[1]
    X[1,:] = x0
    for step in 2:N+1
        T[step] = T[step-1] + h
        k1 = fun(X[step-1,:], p, T[step-1])
        k2 = fun(X[step-1,:] + h*0.5*k1, p, T[step-1]+0.5*h)
        X[step,:] = X[step-1,:] + h * k2
    end

return T,X
end
```

```julia
function heun(fun, tspan::Tuple{<:Real,<:Real}, x0::Vector{<:Real}, p::Vector{<:Real}, N::Integer)
# Numerical integration via the Heun scheme
# the same as Euler
# ----------------------------------------------

    n  = length(x0)             # dimension n
    T  = zeros(N+1)
    X  = zeros(N+1,n)
    h = ((tspan[2]-tspan[1])/N) # step
    
    #Filling X and T
    T[1] = tspan[1]
    X[1,:] = x0
    for step in 2:N+1
        T[step] = T[step-1] + h
        k1 = fun(X[step-1,:], p, T[step-1])
        k2 = fun(X[step-1,:] + h*(1/3)*k1, p, T[step-1]+(1/3)*h)
        k3 = fun(X[step-1,:] + h*(2/3)*k2, p, T[step-1]+(2/3)*h)
        X[step,:] = X[step-1,:] + h * (0.25*k1 + 0.75*k3)
    end
    
return T,X
end
```

```julia
function rk4(fun, tspan::Tuple{<:Real,<:Real}, x0::Vector{<:Real}, p::Vector{<:Real}, N::Integer)
# Numerical integration via the RK4 scheme
# the same as Euler
# ----------------------------------------------

    n  = length(x0)             # dimension n
    T  = zeros(N+1)
    X  = zeros(N+1,n)
    h = ((tspan[2]-tspan[1])/N) # step
    
    #Filling X and T
    T[1] = tspan[1]
    X[1,:] = x0
    for step in 2:N+1
        T[step] = T[step-1] + h
        k1 = fun(X[step-1,:], p, T[step-1])
        k2 = fun(X[step-1,:] + h*0.5*k1, p, T[step-1]+0.5*h)
        k3 = fun(X[step-1,:] + h*0.5*k2, p, T[step-1]+0.5*h)
        k4 = fun(X[step-1,:] + h*k3, p, T[step-1]+1*h)
        X[step,:] = X[step-1,:] + h * ((1/6)*k1 + (2/6)*k2 + (2/6)*k3 + (1/6)*k4)
    end
    
return T,X
end
```

```julia
function vdp(x::Vector{<:Real},p::Vector{<:Real},t::Real)
# Van der Pol model
# second member of the IVP
# p : paramter vector
# t : time, variable not use here

    xpoint = similar(x)
    xpoint[1] = x[2]
    xpoint[2] = (1-(x[1])^2)*x[2] - x[1]
    return xpoint    
end

function affiche(T,X)
# Affiche les résulats : utile pour vérifier que l'on a les résultats attendus pour N=25
#
    N = length(T)-1
    for i in 1:N+1
        @printf("%20.15f %20.15f %20.15f\n", T[i],X[i,1],X[i,2])
    end
end

mu = 1
p = [mu]
tf = 6.6632868593231301896996820305
tspan = (0.0, tf)
x0 = [2.00861986087484313650940188,0]
N = 10

println("Euler")
T_Euler,X_Euler = euler(vdp, tspan, x0, p, N)
affiche(T_Euler,X_Euler)

println("Runge")
T_Runge,X_Runge = runge(vdp, tspan, x0, p, N)
affiche(T_Runge,X_Runge)

println("Heun")
T_Heun,X_Heun = heun(vdp, tspan, x0, p, N)
affiche(T_Heun,X_Heun)

println("RK4")
T_RK4,X_RK4 = rk4(vdp, tspan, x0, p, N)
affiche(T_RK4,X_RK4)
```

## Runge Kutta implicite, schéma de Gauss
### Point fixe
#### Code


```julia
function gauss_fp(fun, tspan::Tuple{<:Real,<:Real}, x0::Vector{<:Real}, p::Vector{<:Real},
        option)::Tuple{Vector{Float64},Array{Float64,2},Int,Array{Int,1}}
# Numerical integration with Gauss scheme and the fixed point algorithm
#
# Input parameters
# ----------------
# fun : second member of the ode
#       fun(x,p,t)
#       Input
#       x      : state
#                array(n)
#       p      : vector of parameters
#       t      : time
#                real
#       Output
#       xpoint : value of fun
#               array(n) 
# tspan = (t0,tf)
# x0    = initial point
#         array(n)
# option[1] = N = number of step
#                 Integer
# option[2] = maximum number of iterations for the fixed point
#             real
# option[3] = epsilon for the test of progress in the fixed point
#             real
#
# Output parameters
# -----------------
# T = vector of time
#     real(N+1)
# X = Matrix of solution
#     real(N+1,n)
# The line i of [T X] contains ti and x_i the approximation of x(t_i)
# ifail[i] = number of iteration for the fixed point on [t_i,t_{i+1}] if the fixed point converge
# ifail[i] = -1 if the fixed point don't converge on [t_i,t_{i+1}]: maximum number of iteration
# is attained in the fixed point
#            integer(N)
# nfun = number of evaluation of phi
#        integer
# ndfun = number of evaluation of the derivative of fun
#         integer
# ---------------------------------------------------------------------------------------------
#
# Initialisation
# --------------
# Coefficent of the Butcher's array
    N = Int(option[1])
    n  = length(x0)             # dimension n
    T  = zeros(N+1)
    X  = zeros(N+1,n)
    nfun = 0
    ifail = ones(Int,N)
    h = ((tspan[2]-tspan[1])/N) # step
    
    #Filling X and T
    T[1] = tspan[1]
    X[1,:] = x0
    
    for step in 2:N+1
        
        T[step] = T[step-1] + h
        
        cptr_ite = 0 #counter of iteration for fixed point algorithm
        norm_step = 1 #norm of the step vector between two iterations in fixed point algorithm  
        
        # First value for k1 and k2
        k1 = fun(X[step-1,:], p, T[step-1]+h*(0.5-sqrt(3)/6))
        k2 = fun(X[step-1,:], p, T[step-1]+h*(0.5+sqrt(3)/6))
        
        nfun = nfun+2 #increment of fun call counter      
        
        #definition of boolean for iterations
        exceed = cptr_ite > option[2]
        convergence = norm_step < option[3]
        
        #Fixed Point iterations
        while (!exceed) && (!convergence)
            next_k1 = fun(X[step-1,:] + h *(0.25*k1 + (0.25-sqrt(3)/6)*k2), p, T[step-1]+h*(0.5-sqrt(3)/6))
            next_k2 = fun(X[step-1,:] + h *((0.25+sqrt(3)/6)*k1 + 0.25*k2), p, T[step-1]+h*(0.5+sqrt(3)/6))
            
            nfun = nfun+2 #increment of fun call counter
            
            norm_step = norm([next_k2-k2,next_k1-k1]) #new norm of the step vector between iterations
            convergence = norm_step <= option[3] #update of convergence boolean
            
            #update of k1 and k2 for next iteration
            k1 = next_k1
            k2 = next_k2
            
            #increment of iteration counter and update of max iterations boolean
            cptr_ite = cptr_ite + 1
            exceed = cptr_ite >= option[2]
        end
        
        #convergence study
        if convergence
            ifail[step-1] = cptr_ite
        else
            ifail[step-1] = -1
        end      
        
        X[step,:] = X[step-1,:] + 0.5*h*(k1+k2)
    end
return T,X,nfun,ifail
end
```

### Newton (non terminé)
#### Code

```julia
function gauss_newton(fun, dfun, tspan::Tuple{<:Real,<:Real}, x0::Vector{<:Real}, p::Vector{<:Real},
        option)::Tuple{Vector{Float64},Array{Float64,2},Int,Int,Array{Int,1}}
# Numerical integration with Gauss scheme and the Newton algorithm
#
# Input parameters
# ----------------
# fun : second member of the ode
#       fun(x,p,t)
#       Input
#       x      : state
#                array(n)
#       p      : vector of parameters
#       t      : time
#                real
#       Output
#       xpoint : value of fun
#               array(n) 
# tspan = (t0,tf)
# x0    = initial point
#         array(n)
# option[1] = N = number of step
#                 Integer
# option[2] = maximum number of iterations for the Newton algorithm
#             real
# option[3] = epsilon for the test of progress in the Newton algorithm
#             real
#
# Output parameters
# -----------------
# T = vector of time
#     real(N+1)
# X = Matrix of solution
#     real(N+1,n)
# The line i of [T X] contains ti and x_i the approximation of x(t_i)
# ifail[i] = number of iteration for the fixed point on [t_i,t_{i+1}] if the fixed point converge
# ifail[i] = -1 if the fixed point don't converge on [t_i,t_{i+1}]: maximum number of iteration
# is attained in the fixed point
#            integer(N)
# nfun = number of evaluation of phi
#        integer
# ndfun = number of evaluation of the derivative of fun
#         integer
# ---------------------------------------------------------------------------------------------

    N = Int(option[1])
    n  = length(x0)             # dimension n
    T  = zeros(N+1)
    X  = zeros(N+1,n)
    nfun = 0
    ndfun = 0
    ifail = ones(Int,N)    
    h = ((tspan[2]-tspan[1])/N) # step
       
    #Filling X and T
    T[1] = tspan[1]
    X[1,:] = x0
    
    #a_{i,j} coefficients matrix
    A = zeros(2,2)
    A[1,1] = 0.25
    A[1,2] = 0.25 - sqrt(3)/6
    A[2,1] = 0.25 + sqrt(3)/6
    A[2,2] = 0.25
    
    # 2x2 identity matrix
    Id = Matrix(1I,2,2)
        
    for step in 2:N+1
        
        T[step] = T[step-1] + h
        
        cptr_ite = 0 #counter of iteration for fixed point algorithm
        norm_step = 1 #norm of the step vector between two iterations in fixed point algorithm  
        
        # First value for Z vector 
        Z = zeros(2,1)
        
        # J matrix for this step
        J = dfun(X[step-1,:], p, T[step-1])
                
        #definition of boolean for iterations
        exceed = cptr_ite > option[2]
        convergence = norm_step < option[3]
        
        #Fixed Point iterations
        while (!exceed) && (!convergence)
            
            F[1] = fun(X[step-1,:]+[Z[1],Z[2]], p, T[step-1]+(0.5-sqrt(3)/6)*h)
            F[1] = fun(X[step-1,:]+[Z[1],Z[2]], p, T[step-1]+(0.5+sqrt(3)/6)*h)

            ΔZ = inv(Id-h*A*dfun(X[step,:],p,T[step]))*((-1)*Z+h*A*F)
            
            next_Z = Z + ΔZ
            
            ndfun = ndfun +1
            nfun = nfun +2 
            
            norm_step = norm(next_Z-Z) #new norm of the step vector between iterations
            convergence = norm_step <= option[3] #update of convergence boolean
            
            #update of k1 and k2 for next iteration
            Z = next_Z
                       
            #increment of iteration counter and update of max iterations boolean
            cptr_ite = cptr_ite + 1
            exceed = cptr_ite >= option[2]
        end
        
        #convergence study
        if convergence
            ifail[step-1] = cptr_ite
        else
            ifail[step-1] = -1
        end      
        
        g1 = X[step-1,:]+[Z[1],Z[1]]
        g2 = X[step-1,:]+[Z[2],Z[2]]
        
        X[step,:] = X[step-1,:] + h*(0.5*fun(g1, p, T[step-1]+h*(0.5-sqrt(3)/6))+0.5*fun(g2, p, T[step-1]+h*(0.5+sqrt(3)/6)))
        
        nfun = nfun +2
    end

return T,X,nfun,ndfun,ifail
end
```

#### Test dans le cas N = 10, nb_itmax = 15 et K_eps = 1.e-12

```julia
function dvdp(x::Vector{<:Real},p::Vector{<:Real},t::Real)::Matrix{<:Real}
# Van der Pol model
# derivative of the second member of the IVP
# x : state
#     real(n)
# p : paramter vector
# t : time, variable not used here
#     real
# Output
# Xpoint::Jacobian matrix{<:Real},
#       real(n,n)
# --------------------------------------
    n = length(x)
    Jac = ones(n,n)
    Jac[1,1] = 0
    Jac[1,2] = 1
    Jac[2,1] = 2*x[1]*x[2]-1
    Jac[2,2] = 1-x[1]^2
    return Jac
end

function vdp(x::Vector{<:Real},p::Vector{<:Real},t::Real)
# Van der Pol model
# second member of the IVP
# p : paramter vector
# t : time, variable not use here

    xpoint = similar(x)
    xpoint[1] = x[2]
    xpoint[2] = (1-(x[1])^2)*x[2] - x[1]
    return xpoint    
end

function affiche(T,X)
# Affiche les résulats : utile pour vérifier que l'on a les résultats attendus pour N=10
#
    N = length(T)-1
    for i in 1:N+1
        @printf("%20.15f %20.15f %20.15f\n", T[i],X[i,1],X[i,2])
    end
end

mu = 1
p = [mu]
tf = 6.6632868593231301896996820305
tspan = (0.0, tf)
x0 = [2.00861986087484313650940188,0]
N = 10
nb_itmax = 15
K_eps = 10^(-12)
options = [N, nb_itmax, K_eps]

println("Gauss with Newton algorithm")
T_Gauss, X_Gauss, nfun, ndfun, ifail = gauss_newton(vdp, dvdp, tspan, x0, p, options)
affiche(T_Gauss,X_Gauss)
println(nfun)
print(ifail)
```
